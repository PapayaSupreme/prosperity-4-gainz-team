# Prosperity 4 Tutorial Round Starter

This notebook gives a quick first pass on spread behavior and price dynamics for `EMERALDS` and `TOMATOES`.

```python
from pathlib import Path
import sys

import pandas as pd
import plotly.express as px

repo_root = Path.cwd().resolve().parents[1] if Path.cwd().name == 'analysis' else Path.cwd().resolve()
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

from prosperity4.analysis.data import read_all_tutorial_prices, read_all_tutorial_trades

prices = read_all_tutorial_prices()
trades = read_all_tutorial_trades()

print(f'Prices shape: {prices.shape}')
print(f'Trades shape: {trades.shape}')
prices.head()
```

```python
df = prices.copy()
df['spread'] = df['ask_price_1'] - df['bid_price_1']
df['mid'] = (df['ask_price_1'] + df['bid_price_1']) / 2

spread_stats = (
    df.groupby('product')['spread']
    .agg(['count', 'mean', 'median', 'std', 'min', 'max'])
    .sort_index()
)
spread_stats
```

```python
sample = (
    df.sort_values(['product', 'timestamp'])
    .groupby('product', as_index=False, group_keys=False)
    .head(3000)
)

fig_mid = px.line(sample, x='timestamp', y='mid', color='product', title='Mid Price (first 3000 points per product)')
fig_mid.show()
```

```python
sample_spread = (
    df.sort_values(['product', 'timestamp'])
    .groupby('product', as_index=False, group_keys=False)
    .head(3000)
)

fig_spread = px.line(sample_spread, x='timestamp', y='spread', color='product', title='Bid/Ask Spread (first 3000 points per product)')
fig_spread.show()
```
